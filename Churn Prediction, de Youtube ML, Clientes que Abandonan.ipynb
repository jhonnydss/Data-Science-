# Scaling 

Para que tengan el mismo nivel de las nicotomicas que esten en 1 y 0 como las de monthlycharge son dineros.

# Seleccionar las X y las y

```python
X = df2.drop('Churn', axis ='columns')
y = df2['Churn']
```

## Train and test split

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X,y,test_size=0.2,random_state=5)
```

```python
# keras es neuralnetworth
# en el modelo podemos introducir layers 1 por 1
# imput layers, es cuantas neuronas queremos tener
# 26 por las columnas
import tensorflow as tf
from tensorflow import keras 


model= keras.Sequential([
    keras.layers.Dense(20, input_shape=(26,),activation= 'relu'),
    keras.layers.Dense(1, activation= 'sigmoid'),
])

model.compile(optimizer='adam',
             loss='binary_crossentropy',
             metrics=['accuracy'])

model.fit(X_train,y_train, epochs=100)
```

```python
model.evaluate(X_test,y_test)
```

```python
yp = model.predict(X_test)
yp[:5]
```

```python
y_test[:10]
```

# Hacer un for loop para convertir los datos 

Si es mas grande que 0.5 entonces 

```python
y_pred = []
for element in yp:
    if element > 0.5:
        y_pred.append(1)
    else:
        y_pred.append(0)
```

```python
y_pred[:10]
```

```python
from sklearn.metrics import confusion_matrix, classification_report 

print(classification_report(y_test,y_pred))
```

```python
import seaborn as sn
cm = tf.math.confusion_matrix(labels=y_test,predictions=y_pred)

plt.figure(figsize=(10,7))
sn.heatmap(cm,annot=True, fmt='d')
plt.xlabel('Predicted')
plt.ylabel("Truth")
```
